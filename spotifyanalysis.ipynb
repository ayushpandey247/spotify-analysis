```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
tracksdf= pd.read_csv("tracks.csv")
tracksdf.head()
```

```python
pd.isnull(tracksdf).sum()
```

```python
sortedf= tracksdf.sort_values('popularity', ascending= True).head(10)
sortedf
```

```python
tracksdf.describe().transpose()
```

```python
most_popular= tracksdf.query('popularity > 90', inplace= False).sort_values('popularity', ascending= False)
most_popular[:10]
```

```python
print(tracksdf.columns.tolist())
```

```python
tracksdf['release_date']
```

```python
tracksdf.set_index("release_date", inplace= True)
tracksdf.index=pd.to_datetime(tracksdf.index)
tracksdf.head()
```

```python
tracksdf[['artists']].iloc[18]
```

```python
tracksdf['duration']=tracksdf['duration_ms'].apply(lambda x: round(x/1000))
tracksdf.drop('duration_ms', inplace= True, axis=1)
tracksdf.head()
```

```python
corrdf= tracksdf.drop(['key','mode','explicit'], axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corrdf, annot= True,fmt=".1g",vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="Black")
heatmap.set_title("Correlation Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sampledf= tracksdf.sample(int(0.004*len(tracksdf)))
print(len(sampledf))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data= sampledf, y="loudness", x="energy", color="c").set(title= "Loudness Vs Energy Correlation" )
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data= sampledf, y="popularity", x="acousticness", color="y").set(title= "Popularity Vs Acousticness Correlation" )
```

```python
tracksdf['dates']=tracksdf.index.get_level_values('release_date')
tracksdf.dates=pd.to_datetime(tracksdf.dates)
years=tracksdf.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2, height=5, kind='hist').set(title="Number Of Songs Per Year")
```

```python
total_dr=tracksdf.duration
fig_dims=(18,7)
fig, ax=plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=years, y=total_dr, ax=ax, errwidth=False).set(title= "Year Vs Duration")
plt.xticks(rotation =90)
```

```python
total_dr=tracksdf.duration
sns.set_style(style="whitegrid")
fig_dims=(10,5)
fig, ax=plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years, y=total_dr, ax=ax).set(title= "Year Vs Duration")
plt.xticks(rotation =60)
```
